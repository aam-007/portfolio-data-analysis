```python
import pandas as pd
from pathlib import Path

data_path = Path("data")
files = list(data_path.glob("*.csv"))

dfs = []

for f in files:
    name = f.stem.split()[0]
    df = pd.read_csv(f)

    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

    df = df[["Date", "Price"]]

    df["Price"] = (
        df["Price"]
        .astype(str)
        .str.replace(",", "")
        .astype(float)
    )

    df = df.rename(columns={"Price": name})

    dfs.append(df)

out = dfs[0]

for df in dfs[1:]:
    out = out.merge(df, on="Date", how="outer")

out = out.sort_values("Date").set_index("Date")

out.to_csv("india_equity_prices.csv")

out.head()
```
