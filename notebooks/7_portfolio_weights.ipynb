# 7 · Portfolio Weights — Final Presentation

Loads `portfolio_weights.csv`, converts to percentages, saves a clean CSV,  
and generates a bar chart to `../charts/`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from pathlib import Path
import sys
sys.path.insert(0, '..')
from utils import ANALYSIS_DIR, CHARTS_DIR, apply_style

apply_style()

analysis_path = Path(ANALYSIS_DIR)
charts_path   = Path(CHARTS_DIR)
charts_path.mkdir(parents=True, exist_ok=True)

weights = pd.read_csv(
    analysis_path / "portfolio_weights.csv",
    usecols=["Asset", "Equal_Weight", "MaxSharpe_Weight", "MinVar_Weight"]
)
weights = weights.set_index("Asset")

# ── Percentage table ──────────────────────────────────────────────────────────
weights_pct = (weights * 100).round(2)
weights_pct.columns = ["Equal Weight (%)", "Max Sharpe (%)", "Min Variance (%)"]

weights_pct.to_csv(analysis_path / "portfolio_construction_percent.csv")
print("Portfolio Weights (%)")
print(weights_pct.to_string())
print(f"\nColumn totals: {weights_pct.sum().to_dict()}")

# ── Chart ─────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))
weights_pct.plot(kind="bar", ax=ax, edgecolor="white")
ax.yaxis.set_major_formatter(mticker.PercentFormatter())
ax.set_title("Portfolio Construction — Weight Allocation (%)")
ax.set_ylabel("Weight (%)")
ax.set_xlabel("Stock")
plt.xticks(rotation=45)
ax.legend(loc="upper right")
fig.tight_layout()
fig.savefig(charts_path / "portfolio_construction_percent.png", dpi=150)
plt.close()

print("\nSaved: analysis/portfolio_construction_percent.csv")
print("Saved: charts/portfolio_construction_percent.png")
```
