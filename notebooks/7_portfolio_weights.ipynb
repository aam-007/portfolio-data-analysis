```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

charts = Path("charts")
charts.mkdir(exist_ok=True)

# Load CSV correctly, ignoring unnamed extra column
weights = pd.read_csv("portfolio_weights.csv", usecols=["Asset","Equal_Weight","Optimized_Weight"])

# Set Stock names as index
weights.set_index("Asset", inplace=True)

# Convert to percentages
weights_pct = (weights * 100).round(2)

# Save cleaned CSV
weights_pct.to_csv("portfolio_construction_percent.csv")
print("Saved CSV: portfolio_construction_percent.csv")

# Plot bar chart for Equal vs Optimized
plt.figure(figsize=(12,6))
weights_pct.plot(kind="bar")
plt.title("Portfolio Construction (% Allocation)")
plt.ylabel("Weight (%)")
plt.xlabel("Stock")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(charts/"portfolio_construction_percent.png")
plt.close()
print("Saved chart: charts/portfolio_construction_percent.png")

# Display table
print(weights_pct)
```
