```python
import pandas as pd
import numpy as np

df = pd.read_csv("india_equity_prices.csv", parse_dates=["Date"], index_col="Date")

returns = df.pct_change().dropna()

log_returns = np.log(df / df.shift(1)).dropna()

cov_matrix = returns.cov()

corr_matrix = returns.corr()

returns.to_csv("india_equity_returns.csv")
log_returns.to_csv("india_equity_log_returns.csv")
cov_matrix.to_csv("india_covariance_matrix.csv")
corr_matrix.to_csv("india_correlation_matrix.csv")

returns.head()
```
